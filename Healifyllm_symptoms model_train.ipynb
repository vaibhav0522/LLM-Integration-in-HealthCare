# dataframe cleaning:

##### Basic cleaning:

```python
col_cleaned = []
for col in df.columns:
    if '_' in col:
        col = col.replace("_",' ')
    col_cleaned.append(col)

df.columns = col_cleaned
display(df.head())
```

```python
label_cleaned = []
for row_idx, row in df.iterrows():
	disease = row[-1]
	if '^' in disease:
		disease = disease.replace('^',' / ')
	if '_' in disease:
		disease = disease.replace("_",' ')
	#print(disease)
	label_cleaned.append(disease)

df['label'] = label_cleaned
display(df.head())
```

##### alternative Deep cleaning. Removing all codes. Removing column header codes:

```python
df.drop(columns=['UMLS:C0032739_'], inplace = True)
```

```python
df.columns.tolist()
```

```python
col_cleaned = []
for col in df.columns:
    if '_' in col:
        col = col.split('_')[1]
        if '^' in col: 
            col = col.split('^')[0]
    col_cleaned.append(col)
    
df.columns = col_cleaned
df.columns.tolist()
```

Removing codes from df['label']:

```python
df['disease_label'] = df['label']
#df['diseases_cleaned'].to_list()
```

```python
for row_idx, row in df.iterrows():
	disease = row[-1]
	if '^' in disease:
		while '^' in disease:
			disease = disease.split('^') 
			for list_idx, data in enumerate(disease): 
				disease[list_idx] = data.split('_')[1]
			disease_cleaned = ' / '.join(disease)
	else: 
		disease_cleaned = disease.split('_')[1]
	#print(disease_cleaned)
	df.loc[row_idx, 'disease_label'] = disease_cleaned

df.drop(columns=['label'], inplace = True)
df['disease_label'].to_list()
```

cleaning complete.

```python
df.to_csv('disease_135_dataset_cleaned.csv', index=False, encoding='utf-8')
```
